```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# list of strings
tech = ['C','C#','C++','Java','JavaScript','Python','Scala','Oracle','SQL Server','MySQL Server','PostgreSQL','MongoDB']
num_jobs = [184,14,24,92,65,51,47,8,16,7,17,18]

# Calling DataFrame constructor on list
# with indices and columns specified
df = pd.DataFrame(list(zip(tech, num_jobs)), columns =['Lang', 'Postings'])
df
```

```python
df2 = df.sort_values(by=['Postings'], ascending=False).reset_index(drop=True)
```

```python
df2
```

```python
fig = plt.figure(figsize = (10, 5))
plt.barh(df2['Lang'],df2['Postings'])

plt.xlabel("Number of Job Postings")
plt.ylabel("Language")
plt.title("Number of Job Postings for each Language")
plt.show()
```

```python
plt.figure(figsize=(14, 10))
sns.barplot(
    x="Postings", 
    y="Lang", 
    data=df2, 
    estimator=sum, 
    ci=None, 
    color='#69b3a2')
```

```python
plt.figure(figsize=(14, 10))
ranks = df2.Postings.sort_values().index
palette = sns.color_palette('Blues_d', len(ranks))
ax = sns.barplot(x='Postings', y='Lang', palette=np.array(palette)[ranks], data=df2)
```

```python
def show_values(axs, orient="v", space=.01):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
                value = '{:.1f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.1f}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

```python
plt.figure(figsize=(14, 10))
plt.xlabel("Number of Job Postings")
plt.ylabel("Language")
plt.title("Number of Job Postings for each Language")
p = sns.barplot(x="Postings", y="Lang", data=df2, ci=None)

#show values on barplot
show_values(p, "h", space=0)
```

```python
# list of strings
tech = ['Python','Java','R','Javascript','Swift','C++','C#','PHP','SQL','Go']
pay = [114.383,101.013,92.037, 110.981, 130.801, 113.865, 88.726, 84.727, 84.793, 94.082]

# Calling DataFrame constructor on list
# with indices and columns specified
df3 = pd.DataFrame(list(zip(tech, pay)), columns =['Lang', 'Payment'])
df3
```

```python
df4 = df3.sort_values(by=['Payment'], ascending=False).reset_index(drop=True)
```

```python
df4
```

```python
plt.figure(figsize=(14, 10))
plt.xlabel("Payment in $")
plt.ylabel("Language")
plt.title("Payment in $ for each Language")
p = sns.barplot(x="Payment", y="Lang", data=df4, ci=None)

#show values on barplot
show_values(p, "h", space=0)
```
